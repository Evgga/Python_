```python
import pandas as pd
```

```python
import random
lst = ['robot']*10
lst += ['human']*10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data
```

```python
data.loc[data['whoAmI'] == 'human','human'] = '1'
```

```python
data.loc[data['whoAmI'] != 'human','human'] = '0'
```

```python
data.drop(columns = 'whoAmI')
```
